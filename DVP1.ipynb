HISTOGRAM

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/tips.csv')
plt.hist(data['tip'])
plt.title("HISTOGRAM")
```

BAR CHART

```python
import matplotlib.pyplot as plt
plt.bar(data['sex'],data['tip'])
plt.title('Bar Chart')
plt.xlabel('sex')
plt.ylabel('tip')
plt.show()
```

SCATTER PLOT

```python
import matplotlib.pyplot as plt
plt.scatter(data['sex'],data['tip'])
plt.title('Scatter Plot')
plt.xlabel('sex')
plt.ylabel('tip')
plt.show()
```

LINE CHART USING ARRAY

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,4,5,7,8])
y=x*2
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,8])
y=np.array([3,8])
plt.plot(x,y)
plt.show()
```

BOX PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/tips.csv')
sns.barplot(x='day',y='tip',data= data,hue='sex')
plt.show()
```
